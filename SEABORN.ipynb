```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_excel("cardata.xlsx")
df.head()
```

```python
variables =['Min Price (Lakh)','Max Price (Lakh)', 	'Range (kmpl)', 'CC' ,'Seats', 'Variants', 'Type', 	'Ex-Showroom Price', 'RTO', 'Insurance', 
            'Other','Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i,var in enumerate(variables,start=1):
    plt.subplot(4,3,i)
    sns.scatterplot(x=df[var],y=df['Max Price (Lakh)'],alpha=0.7)
    plt.title(f'scatter plot of{var} vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i,var in enumerate(variables,start=1):
    plt.subplot(4,3,i)
    sns.histplot(df[var],bins=20,kde =True,color='blue')
    plt.title(f'histogram of{var} ')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
