```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =pd.read_excel("Excelpython.xlsx")
df.head()
```

```python
df['Date'] =pd.to_datetime(df['Date'])
plt.plot(df['Date'], df['Sales'],color= "Blue")
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df['Product'], df['Sales'],color= "indigo")
plt.title('Sales per product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
df =pd.read_excel("Excelpython.xlsx")
df.head()
```

```python
sales_data =df.groupby('Category')['Sales'].sum()
sales_data
```

```python
plt.figure(figsize=(4,4))
plt.pie(sales_data,labels=sales_data.index,autopct='%1.1f%%')
plt.title('Proportion of Sales by Category')
plt.show()
```

```python
grouped_data =df.groupby('Region')['Sales'].sum()
grouped_data
```

```python
plt.figure(figsize=(5,5))
plt.pie(grouped_data,labels=grouped_data.index,autopct='%1.3f%%')
plt.title('Proportion of Sales by Category')
plt.show()
```

```python
plt.hist(df['Sales'], bins=8,color='blue',edgecolor='purple')
plt.title('Histogram')
plt.xlabel('Sales bins')
plt.ylabel('frequency')
plt.show()
```

```python
plt.fill_between(df['Date'],df['Sales'],color='black',alpha=0.3)
plt.plot(df['Date'], df['Sales'],color= "Blue")
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.fill_between(df['Product'],df['Sales'],color='indigo',alpha=0.3)
plt.bar(df['Product'], df['Sales'],color= "Blue")
plt.title('Sales Over Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
